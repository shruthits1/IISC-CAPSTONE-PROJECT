## Top losers, gainers and actively traded

```python
url = 'https://www.alphavantage.co/query?function=TOP_GAINERS_LOSERS&apikey=demo'
r = requests.get(url)
data = r.json()
print(data)
df = pd.DataFrame(data['most_actively_traded'])
print(df.head())
```

```python
top50tickers = df['ticker'].tolist()
print(top50tickers)
```

## Get info for top 50 tickeers

```python
def get_info_api(url):
  r = requests.get(url)
  data = r.json()
  print(data)
  return data
```

```python
all_data = []
for ticker in top50tickers:
    url = 'https://www.alphavantage.co/query?function=OVERVIEW&symbol=' + ticker + '&apikey=demo'
    data = get_info_api(url)
    # Check if the data is valid before converting to DataFrame
    if data and 'Symbol' in data:
        df_ticker = pd.DataFrame.from_dict(data, orient='index').T
        all_data.append(df_ticker)
    else:
        print(f"Could not retrieve data for ticker: {ticker}")

# Concatenate all DataFrames in the list
if all_data:
    df = pd.concat(all_data, ignore_index=True)
    display(df)
else:
    print("No data was retrieved for any ticker.")
```
